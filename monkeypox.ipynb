```python
def top_10(col=None):
  #sort dset
  df2_sort = df2.sort_values(by=col, ascending=False).reset_index()
  #take top 10
  top10 = df2_sort[:10]
  #plot top 10
  label = ''.join(col.split('_'))
  labeldict = {'size':'15', 'weight':'3'}
  titledict = {'size':'20', 'weight':'3'}
  fig = px.bar(x='Country',
               y=col,
               data_frame=top10,
               labels = ['Country', label],
               color = col,
               color_continuous_scale = 'electric',
               text_auto = True,
               title = f'Top 10 Countries based on {label}')
  fig.show()
```

```python
top_10(col='Confirmed_Cases')
```

```python
def world_map(col=None, title=None):
  '''
  Fx to plot a world map
  Args required:
  1.  Col name for which distribution is to be plotted
  2.  Graph title
  '''
  fig = px.choropleth(df2,
                      locations='Country',
                      locationmode='country names',
                      color=col,
                      color_continuous_scale='Viridis')
  
  fig.update_layout(title_text=title)
  fig.show()
```

```python
world_map(col='Confirmed_Cases', title='World confirmed monkeypox cases')
```
